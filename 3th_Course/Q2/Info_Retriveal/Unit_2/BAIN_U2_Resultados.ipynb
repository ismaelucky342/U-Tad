# Ejercicio 1: Limpieza y Tokenización

Dado un conjunto de textos, crear una función que:
- Convierta el texto a minúsculas.
- Elimine la puntuación.
- Tokenice el texto usando NLTK.
- Asigna el resultado a una nueva columna en un DataFrame de Pandas.

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt-tab')

data = {"texto": ["¡Hola Mundo! Este es un Ejemplo de Minería de Texto.",
                  "Los datos se limpian y se tokenizan."]}

df = pd.DataFrame(data)

def limpiar_tokenizar(texto):
    texto = texto.lower()
    texto = re.sub(r"[^\w\s]", "", texto)
    return word_tokenize(texto)

df["tokens"] = df["texto"].apply(limpiar_tokenizar)
print(df)
```

# Ejercicio 2: Expresiones Regulares

Usar expresiones regulares para extraer todos los números (incluyendo decimales) de un texto dado.

```python
texto = "El precio es 45.67 USD y el descuento es 10.5%."

import re
patron = r"\d+(?:[.,]\d+)?"
numeros = re.findall(patron, texto)
print("Números extraídos:", numeros)
```

# Ejercicio 3: Procesamiento con spaCy

Cargar un texto, procesarlo con spaCy y extraer:
- Tokens
- Etiquetas PoS
- Entidades nombradas (NER)

```python
pip install es_core_news_sm
```

```python
import spacy
texto = "La U-Tad organiza un evento en Madrid con expertos en NLP."

nlp = spacy.load("es_core_news_sm")
doc = nlp(texto)
tokens = [token.text for token in doc]
pos_tags = [token.pos_ for token in doc]
entidades = [(ent.text, ent.label_) for ent in doc.ents]
print("Tokens:", tokens)
print("PoS:", pos_tags)
print("Entidades:", entidades)
```

# Ejercicio 4: Visualización con WordCloud

A partir de un DataFrame con textos (DataFrame del ejercicio 2), calcular la frecuencia de palabras y generar una wordcloud usando la librería WordCloud

```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Usamos el DataFrame del Ejercicio 1
all_tokens = [token for tokens in df["tokens"] for token in tokens]
freq = Counter(all_tokens)
wc = WordCloud(width=800, height=600, background_color="white").generate_from_frequencies(freq)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
